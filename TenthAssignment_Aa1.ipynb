```python
#Problem 1
import matplotlib.pyplot as plt
%matplotlib inline
csv_path = "/content/drive/MyDrive/mtfuji_data (1).csv" 

np.set_printoptions(suppress=True) 
fuji = np.loadtxt(csv_path, delimiter=",", skiprows=1)
  
x = []
y = []
  

         
for row in fuji:
  x.append(int(row[0]))
  y.append(int(row[3]))
  

plt.xlabel('Number of point')
plt.ylabel('Elevation')
plt.title('Mt.Fuji')
plt.plot(x, y)
plt.show()
```

```python
#Problem 2
def calc_slope(point):
  
  change_in_x = fuji[:,0][point] - fuji[:,0][point-1]
  change_in_y = fuji[:,3][point] - fuji[:,3][point-1]
  slope = round(change_in_y//change_in_x)
  
  #print(slope)
  return slope
  

slope = calc_slope(136)
```

```python
#Problem 3
def calc_dest(current_point, alpha):
  #slope = calc_slope(current_point)
  dest = (round(current_point - alpha*slope))
  if (dest <0):
    return 0 
  else:
    return dest

calc_dest(136, 0.2)
```

```python
#Problem 4
def going_down(current_point, alpha):
  
  #looping until the returned destination = 0
  
  destinations = [current_point]
  for counter in range(fuji.size):
    current_point = round(calc_dest(current_point, alpha))
    if(destinations[-1] == current_point):
      break
    else:
      destinations.append(current_point)
  return destinations
  
going_down(136, 0.2)
```

```python
#Problem 5
%matplotlib inline
import matplotlib.pyplot as plt


positions_array = going_down(136,0.2)

for counter in range(len(positions_array)):
  changes_xq = (fuji[:,0][positions_array]) - (fuji[:,0][positions_array[-1]])
  changes_yq = (fuji[:,3][positions_array]) - (fuji[:,3][positions_array[-1]])
  t = changes_yq//changes_xq
print(t)
                         
figure = plt.figure(figsize=(15,5))

fig1 = figure.add_subplot(1,2,1)

fig1.set_title("MT. Fuji movement points")
fig1.set_xlabel("position")
fig1.set_ylabel("elevation(m)")
fig1.plot(fuji[:,0],fuji[:,3])
fig1.scatter(positions_array,[fuji[position_2, 3] for position_2 in positions_array],c="orange")

fig2 = figure.add_subplot(1,2,2)

fig2.set_title("MT. Fuji movement points")
fig2.set_xlabel("Gradient")
fig2.set_ylabel("Elevation (m)")
fig2.plot(t, fuji[:,3][positions_array])
```
