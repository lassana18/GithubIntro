```python
#problem 1
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
np.random.seed(0)
mean = [-3, 0]
cov = [[1.0, 0.8], [0.8, 1.0]]
size = [500]
M = np.random.multivariate_normal(mean, cov, size) 
M.shape
```

```python
#Problem 2
x= M[:,0]
y = M[:,1]
plt.title("Scater Plot")
plt.xlabel("First Columnn")
plt.ylabel("Second Column")
plt.show
plt.scatter(x, y)
```
