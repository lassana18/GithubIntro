```python
#Problem 1
import matplotlib.pyplot as plt
%matplotlib inline
csv_path = "/content/drive/MyDrive/mtfuji_data (1).csv" 

np.set_printoptions(suppress=True) 
fuji = np.loadtxt(csv_path, delimiter=",", skiprows=1)
  
x = []
y = []
  

         
for row in fuji:
  x.append(int(row[0]))
  y.append(int(row[3]))
  

plt.xlabel('Number of point')
plt.ylabel('Elevation')
plt.title('Mt.Fuji')
plt.plot(x, y)
plt.show()
```

```python
#Problem 2
x_np = np.array(x)
y_np = np.array(y)
change_x = x_np[1:] - x_np[:-1] 
change_y = y_np[1:] - y_np[:-1] 
gradient = change_y/change_x 
Gradient1 = np.append(gradient, [0])
print(gradient)
print(Gradient1.size)
```
